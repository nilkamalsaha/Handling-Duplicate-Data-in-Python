## 4. Duplicate Data Check

* Find any duplicate data
* Deal with the duplicate data

```python
import pandas as pd
alarm_survey = pd.read_excel('Alarm Survey Data.xlsx')

#duplicate rows in the DataFrame and returns a Boolean Series
alarm_survey.duplicated()
```

```python
#return all duplicate rows

alarm_survey[alarm_survey.duplicated(keep=False)]
```

```python
# Removes duplicate rows from the alarm_survey DataFrame and Modifies the original DataFrame in place

alarm_survey.drop_duplicates(inplace = True)
```

```python
alarm_survey
```

```python
# iloc: Stands for index location. It is used for purely integer-based indexing.
# 40:45: This slices rows from index 40 to 44 (Python slicing is exclusive of the end, so it stops before 45).
# : Selects all columns.

alarm_survey.iloc[40:45, :]
```

```python
alarm_survey.reset_index(drop = True, inplace = True)
```
